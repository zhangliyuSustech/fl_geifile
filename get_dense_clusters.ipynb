```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.cluster import KMeans
```

```python
from meshlonlat import mesh2lonlat, lonlat2mesh
```

```python
from collections import Counter
```

```python
df_weekday = pd.read_csv('/data/fan/data/UsersInOsaka_2012/20121101.csv', header=None)
df_weekend = pd.read_csv('/data/fan/data/UsersInOsaka_2012/20121103.csv', header=None)
```

```python
points = np.concatenate([df_weekday[[2, 3]].values, df_weekend[[2, 3]].values], axis=0)
```

```python
points.shape
```

```python
mesh_cnt = Counter()

for point in points:
    mesh = lonlat2mesh(point[1], point[0], meshsize=1000)
    mesh_cnt.update({mesh: 1})
```

```python
len(mesh_cnt)
```

```python
freq_mesh = mesh_cnt.most_common(1600)
```

```python
anchor_points = []
for mesh, _ in freq_mesh:
    anchor_points.append(mesh2lonlat(mesh, meshsize=1000, is_center=True))
```

```python
anchor_points
```

```python
from sklearn.neighbors import KDTree

import pickle as pk

kdtree = KDTree(anchor_points)

with open('anchor_points_osaka.pk', 'wb') as f:
    pk.dump(kdtree, f)
```

```python
from sklearn.neighbors import KDTree
import pickle as pk

with open('./anchor_points_tokyo.pk', 'rb') as f:
    kdtree = pk.load(f)
```

```python
import numpy as np
```

```python
from geopy.distance import geodesic
```

```python
anchor_points = np.array(kdtree.data)
```

```python
min_dists = []

for i in range(anchor_points.shape[0]):
    min_dist = np.inf
    for j in range(anchor_points.shape[0]):
        if i != j:
            min_dist = min(geodesic(anchor_points[i], anchor_points[j]).kilometers, min_dist)
    min_dists.append(min_dist)
```

```python
min_dists = np.array(min_dists)
print(min(min_dists))
print(max(min_dists))
print(np.median(min_dists))
print(np.mean(min_dists))
```
