```python
def filter_and_average(results, dtype=None):
    metrics = []
    for dt in results:
        if len(dt) == 3:
            if dtype is None:
                metrics.append(results[dt])
            else:
                metrics.append(results[dt].item())
    metrics = np.array(metrics)
    return np.mean(metrics, axis=0)
```
