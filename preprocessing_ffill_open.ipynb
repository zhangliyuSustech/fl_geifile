```python
import pickle as pk
import numpy as np
import os
from meshlonlat import lonlat2mesh, mesh2lonlat
from collections import Counter
```

```python
mesh_cnt = Counter()
```

```python
for filename in os.listdir('../data/foursquare/'):

    
    data_interp = dict({})

    fullpath = '../data/foursquare/' + filename
    print(fullpath)
    
    with open(fullpath, 'rb') as f:
        data = pk.load(f)

    for uid in data:
        lat = data[uid][0][1]
        lon = data[uid][0][2]
        mesh = lonlat2mesh(lon, lat, meshsize=1000)
        mesh_cnt.update({mesh: 1})
```

```python
freq_mesh = mesh_cnt.most_common(128)
```

```python
anchor_points = []
for mesh, _ in freq_mesh:
    anchor_points.append(mesh2lonlat(mesh, meshsize=1000, is_center=True))
```

```python
from sklearn.neighbors import KDTree

import pickle as pk

anchor_points = KDTree(anchor_points)

with open('anchor_points_open.pk', 'wb') as f:
    pk.dump(anchor_points, f)
```

```python
T = 96
dT = 900
```

```python
for filename in os.listdir('../data/foursquare/'):

    data_interp = dict({})

    fullpath = '../data/foursquare/' + filename
    print(fullpath)

    with open(fullpath, 'rb') as f:
        data = pk.load(f)

    for uid in data:
        t_beg = int(data[uid][0][0]) // dT
        lat = data[uid][0][1]
        lon = data[uid][0][2]
        loc = anchor_points.query([[lat, lon]], return_distance=False)[0, 0]
        data_interp[uid] = np.zeros(T, dtype=np.int) + loc

        for i in range(1, len(data[uid])):
            t_beg = int(data[uid][i][0]) // dT
            lat = data[uid][i][1]
            lon = data[uid][i][2]
            loc = anchor_points.query([[lat, lon]], return_distance=False)[0, 0]
            data_interp[uid][t_beg:] = loc

    with open('../data/foursquare_interp_128/' + filename, 'wb') as f:
        pk.dump(data_interp, f)
```
