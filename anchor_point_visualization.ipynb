```python
import folium as fl
```

```python
import pickle as pk
```

```python
import numpy as np
```

```python
fmap = fl.Map(location=(35.5, 139.5))
```

```python
with open('./anchor_points.pk', 'rb') as f:
    anchor_points = pk.load(f)
```

```python
anchor_points = np.array(anchor_points.data)
```

```python
for point in anchor_points:
    fl.Marker(list(point)).add_to(fmap)
```
