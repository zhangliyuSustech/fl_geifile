```python
import csv
import numpy as np
import pickle as pk
```

```python
import os
```

```python
import time
```

```python
for m in range(1, 3):
    for d in range(1, 32):
        
        filename = '/data/fan/data/UsersInOsaka_2012/2013{:02d}{:02d}.csv'.format(m, d)
        if not os.path.isfile(filename):
            continue
        print(filename)
        data = dict({})
        start_time = time.mktime(time.strptime('2013-{:02d}-{:02d}'.format(m, d), '%Y-%m-%d')) - 600
        with open(filename, 'r') as f:
            for row in csv.reader(f):
                uid = int(row[0])
                tstamp = time.mktime(time.strptime(row[1], '%Y-%m-%d %H:%M:%S')) - start_time
                if tstamp < 0 or tstamp > 24 * 3600:
                    continue
                lat = float(row[2])
                lon = float(row[3])
                if uid not in data:
                    data[uid] = [(tstamp, lat, lon)]
                else:
                    data[uid].append((tstamp, lat, lon))
                    
            print('Sorting')
            for uid in data:
                data[uid] = sorted(data[uid], key=lambda x: x[0])
                
            with open('/data/fan/UsersInOsakaProcessed/2013{:02d}{:02d}.pk'.format(m, d), 'wb') as f:
                pk.dump(data, f)
```
